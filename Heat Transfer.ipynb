```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
L = 1.0  # Length of the rod
T = 100.0  # Total time
alpha = 0.05  # Thermal diffusivity
N = 100  # Number of grid points
dx = L / (N - 1)  # Grid spacing
dt = 0.001  # Time step
```

```python
x = np.linspace(0, L, N)  # Spatial grid
T_initial = np.zeros(N)  # Initial temperature distribution
T_new = T_initial.copy()  # Temperature at current time step
T_old = T_initial.copy()  # Temperature at previous time step
```

```python
t = 0.0  # Initial time
while t < T:
    # Boundary conditions
    T_new[0] = 0.0  # Left boundary condition
    T_new[N - 1] = 100.0  # Right boundary condition

    # Finite difference scheme
    for i in range(1, N - 1):
        T_new[i] = T_old[i] + alpha * dt / dx**2 * (T_old[i + 1] - 2 * T_old[i] + T_old[i - 1])

    # Update temperature arrays
    T_old = T_new.copy()

    # Increment time
    t += dt
```

```python
plt.plot(x, T_new)
plt.xlabel('Position')
plt.ylabel('Temperature')
plt.title('Heat Transfer Simulation')
plt.show()
```
